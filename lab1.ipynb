```python
#imports
import numpy as np
import csv
import matplotlib.pyplot as plt
```

```python
#function that using for reading data from files
def read_csv_file(file_path):
    with open(file_path) as csvfile:
        csv_reader = csv.reader(csvfile)
        next(csv_reader)
        rows_arr = [[float(col) for col in row[1:]] for row in csv_reader]
    return rows_arr
```

```python
#reading data from train file and visualize it
rows = read_csv_file("lab_1_train.csv")
xColumn = [float(row[0]) for row in rows]
yColumn = [float(row[1]) for row in rows]
plt.title("Train dataset")
plt.plot(xColumn, yColumn, "bo")
plt.show()
```

2. After each epoch, output the current values of weights and loss functions. Use Mean Squared Error as a loss function.

```python
def linear_regression(b0, b1, x):
    return b0 + b1 * x
```

```python
def MSE_loss_function(predicted_row, actual_row):
    return np.mean(([np.square(p - a) for p, a in zip(predicted_row, actual_row)]))
```

```python
#implementation of the gradient descent algorithm to train a linear regression model
def train(rows_arr):
    b0 = float(np.random.randn(1))
    b1 = float(np.random.randn(1))

    x_values = [float(row[0]) for row in rows_arr]
    y_values = [float(row[1]) for row in rows_arr]

    predicted_y = [linear_regression(b0, b1, x) for x in x_values]
    loss = MSE_loss_function(predicted_y, y_values)
    delta = loss

    tolerance = 0.000001
    learning_rate = 0.01
    epoch = 0

    while tolerance < delta:
        predicted_y = [linear_regression(b0, b1, x) for x in x_values]
        difference_y_values = [(expected - predicted) for (expected, predicted) in zip(y_values, predicted_y)]
        derivative_beta0 = -2 * np.mean(difference_y_values)
        sum_val = sum([x * y for x, y in zip(x_values, difference_y_values)])
        derivative_beta1 = -2 * (sum_val / len(difference_y_values))
        b0 = b0 - learning_rate * derivative_beta0
        b1 = b1 - learning_rate * derivative_beta1
        predicted_y = [linear_regression(b0, b1, x) for x in x_values]
        current_loss = MSE_loss_function(predicted_y, y_values)
        delta = loss - current_loss
        loss = current_loss
        print(f"epoch №{epoch}: b0 = {b0}; b1 = {b1}; MSE loss function value = {loss};\n")
        epoch = epoch + 1

    return {
        "b0": b0,
        "b1": b1
    }
```

```python
b_object = train(rows)
```

3. After the gradient descent has converged, calculate the loss function on the test data from lab1_test.csv

```python
test_rows = read_csv_file("lab_1_test.csv")
test_x_values = [float(row[0]) for row in test_rows]
test_y_values = [float(row[1]) for row in test_rows]
predicted_y1 = [linear_regression(b_object["b0"], b_object["b1"], x) for x in test_x_values]
loss_test = MSE_loss_function(predicted_y1, test_y_values)
print(loss_test)
```
